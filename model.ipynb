we combine mother occupation and father occupation to parent_emplpyability risk

wk if either mom or dad occupation is 0, theyre unemployed, >0 means they are employed

1. 0,0--> high risk
2. 0,1 or 1,0--> medium risk
3. 1,2--> low risk

```python
## CHECK IF MOTHER IS EMPLPLOYED
data['Mother_employed'] = (data["Mother's occupation"] != 0).astype(int)
```

```python
## CHECK IF FATHER IS EMPLOYED
data['father_employed'] = (data["Father's occupation"] != 0).astype(int)
```

```python
data['parent_employability_risk'] = 2 - (
    data['Mother_employed'] + data['father_employed']
)
```
